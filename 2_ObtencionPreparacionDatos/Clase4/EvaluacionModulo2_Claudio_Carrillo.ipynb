```python
''' En cada una de las palabras que compone el texto de entrada, tomar cada uno de los caracteres de la palabra y rotarla en 13 posiciones del abecedario. Es decir, si se toma como base el siguiente abecedario: ABCDEFGHIJKLMNÑOPQRSTUVWXYZ, entonces el abecedario rotado será: NÑOPQRSTUVWXYZABCDEFGHIJKLM. Por lo tanto, si a nuestro algoritmo entra la palabra HOLA entonces la salida debería ser TBXN'''
```

```python
#Encriptado
def encriptar(texto, desplazamiento):
    abecedario = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
    abecedario_rotado = abecedario[desplazamiento:] + abecedario[:desplazamiento]
    tabla_traduccion = str.maketrans(abecedario, abecedario_rotado)
    
    palabras = texto.split()
    palabras_encriptadas = []
    
    for palabra in palabras:
        palabra_encriptada = palabra.upper().translate(tabla_traduccion)
        palabras_encriptadas.append(palabra_encriptada)
    
    return ' '.join(palabras_encriptadas)   
```

```python
#Desencriptado
def desencriptar(texto_encriptado, desplazamiento):
    abecedario = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
    abecedario_rotado = abecedario[desplazamiento:] + abecedario[:desplazamiento]
    tabla_traduccion = str.maketrans(abecedario_rotado, abecedario)
    
    palabras_encriptadas = texto_encriptado.split()
    palabras_desencriptadas = []
    
    for palabra_encriptada in palabras_encriptadas:
        palabra_desencriptada = palabra_encriptada.translate(tabla_traduccion)
        palabras_desencriptadas.append(palabra_desencriptada)
    
    return ' '.join(palabras_desencriptadas)
```

```python
#Prueba encriptado
texto_original = "Hola HOLA H4L1 - Hola como te va"
desplazamiento = 13
texto_encriptado = encriptar(texto_original, desplazamiento)
print("Texto encriptado:", texto_encriptado)
```

```python
#Prueba desencriptado
texto_desencriptado = desencriptar(texto_encriptado, desplazamiento)
print("Texto desencriptado:", texto_desencriptado)
```

# Ciclos iterativos

- Sentencia FOR
- Sentencia WHILE
- Función RANGE
- Métodos convenientes

## 1. Sentencia FOR

```python
seq = [1,2,3,4,5]
```

```python
for valor in seq:
    print(valor)
```

```python
animales = ['Leon','Zebra','Cocodrilo','Jirafa']
```

```python
i = 0
for animal in animales:
    print( "Iteracion " + str(i) + ' -> ' +animal)
    i = i + 1
```

```python
seq
```

```python
for val in seq:
    print(val)
```

## 2. Ciclo While

```python
i = 0
while i < 5:
    print( 'Hola ' + str(i) )
    i = i + 1
```

## 3. Funcion Range

```python
range(5)
```

```python
list(range(5))
```

```python
for i in range(8):
    print(i)
```

```python
for i in range(0,10,2):
    print(i)
```

```python
list(range(0,10,2))
```

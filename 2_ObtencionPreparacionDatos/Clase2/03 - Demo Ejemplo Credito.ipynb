# Actividad Aprobación Crédito

Usted está encargado del área de créditos de consumo, y para que todos sus ejecutivos utilicen el mismo criterio, ha decidido crear una función con las reglas de negocio para la aprobación de un crédito. Las reglas para aprobar un crédito son las siguientes:

- Ser mayor de 18 años
- Renta superior a 400 mil
- El monto del crédito no debe exceder 3 veces su renta
- si es mayor de 65, credito no exceda 2 veces su renta
 
La función deberá recibir los parámetros requeridos y deberá retornar un valor booleano **True** o **False** dependiendo si se le otorga el crédito o no.

```python
def evaluar_credito(edad: int, ingreso: float, monto: float) -> bool:
    if edad is None or ingreso is None or monto is None:
        return False
    if edad < 18 or edad < 65 and ingreso >= 400000 and monto <= ingreso * 3:
        return True
```

```python
print(evaluar_credito(18,1_000_000, 2_000_000))
```

```python
# casos que vamos a aprobar
print(evaluar_credito(30,.4e6,1e6))
print(evaluar_credito(70,.8e6,1e6))

# casos que vamos a rechazar
# caso 1: cuando no cumple con la edad
print(evaluar_credito(16,500_000,200_000))

# caso 2: cuando el monto es superior a 3 veces su renta
print(evaluar_credito(26,500_000,3_000_000))

# caso 3: renta inferior a 400000
print(evaluar_credito(26,370_000,3_000_000))

# caso 4: cuando es mayor de 65 y monto superior a 
# dos veces su renta
print(evaluar_credito(76,500_000,1_200_000))

# caso 5: cuando es mayor de 65 y renta menor a 400000
print(evaluar_credito(76,300_000,400_000))
```

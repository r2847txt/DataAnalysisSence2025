# Módulos

```python
%%file mymodule.py
"""
Ejemplo de un modulo Python. Contiene una variable llamada my_variable,
una funcion llamada my_function, y una clase llamada MyClass.
"""

my_variable = 0

def my_function():
    """
    Funcion ejemplo
    """
    return my_variable
    
class MyClass:
    """
    Clase ejemplo
    """

    def __init__(self):
        self.variable = my_variable
        
    def set_variable(self, new_value):
        """
        setea self.variable a un nuevo valor
        """
        self.variable = new_value
        
    def get_variable(self):
        return self.variable
```

```python
import mymodule
```

```python
import mymodule as m
```

```python
m.my_variable
```

```python
m.my_function()
```

```python
miclase = mymodule.MyClass() 
miclase.set_variable(5)
miclase.get_variable()
```

### Módulo Statistic

```python
import statistics
```

```python
print(statistics.mean([-11, 5.5, -3.4, 7.1, -9, 22]))
```

```python
print(statistics.stdev([1, 30, 50, 100]))
```

```python
print(statistics.median([-11, 5.5, -3.4, 7.1, -9, 22]))
```

```python
print(statistics.mode(['red', 'green', 'blue', 'red']))
```

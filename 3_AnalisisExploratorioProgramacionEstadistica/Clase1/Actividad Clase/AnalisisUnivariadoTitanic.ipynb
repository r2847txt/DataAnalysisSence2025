# Análisis univariado de la edad

```python
# Cálculo de medidas de tendencia central para la columna 'Age'
media_edad = df['Age'].mean()
mediana_edad = df['Age'].median()
moda_edad = df['Age'].mode()

print(f"Media de la edad: {media_edad}")
print(f"Mediana de la edad: {mediana_edad}")
print(f"Moda de la edad: {moda_edad}")
```

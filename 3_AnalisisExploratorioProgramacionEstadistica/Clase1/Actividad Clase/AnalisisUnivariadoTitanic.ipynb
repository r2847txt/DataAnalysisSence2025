# Análisis univariado de la edad

```python
# Cálculo de medidas de tendencia central
media_edad = df['Age'].mean()
mediana_edad = df['Age'].median()
moda_edad = df['Age'].mode()

print(f"Media de la edad: {media_edad}")
print(f"Mediana de la edad: {mediana_edad}")
print(f"Moda de la edad: {moda_edad}")
```

```python
# Cálculo de medidas de dispersión
rango_variacon_edad = df['Age'].max() - df['Age'].min()
varianza_edad = df['Age'].var()
desviacion_estandar_edad = df['Age'].std()

print(f"Rango de variación de la edad: {rango_variacon_edad}")
print(f"Varianza de la edad: {varianza_edad}")
print(f"Desviación estándar de la edad: {desviacion_estandar_edad}")

# NOTA: para la varianza se utilizo el metodo .var() ya que el dataset es una muestra (891 registros) y no la poblacion completa (2208 registros).
```

```python
# Cálculo de medidas de posición
q1 = df['Age'].quantile(q=0.25)
q2 = df['Age'].quantile(q=0.50)
q3 = df['Age'].quantile(q=0.75)

print('Q1:',q1)
print('Q2:',q2)
print('Q3:',q3)
```
